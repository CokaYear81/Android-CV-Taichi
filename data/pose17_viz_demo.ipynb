# pose17_v1 Data Viewer

Change only `INPUT_JSON` below to visualize another sample in this folder.

```python
from pathlib import Path
import sys

DATA_DIR = Path.cwd()
ANDROID_APP_ROOT = DATA_DIR.parent
if str(ANDROID_APP_ROOT) not in sys.path:
    sys.path.insert(0, str(ANDROID_APP_ROOT))

INPUT_JSON = "0004_firstaction1_20260408_202501.json"
ROTATION = "clockwise_90"
```

```python
from pose17_viz import load_sample, summarize_sample, preview_frames, animate_sample
```

```python
sample_path = DATA_DIR / INPUT_JSON
sample = load_sample(sample_path)
summary = summarize_sample(sample)
summary
```

```python
preview_frames(sample, rotation=ROTATION)
```

```python
animate_sample(sample, rotation=ROTATION, interval_ms=33, frame_step=1)
```
